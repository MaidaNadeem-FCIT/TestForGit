```python
#  b) Assigning Labels after loading Data.....
df2= pd.read_csv("datasets/groupdatawithoutcollables.csv", header=None)
df2.head(4)
```

```python
df2.columns=["Roll No", "Name", "Age", "Address", "Session", "Group","Gender", "Subj 1", "Subj 2", "Scholarship"]
df2.head(5)
```

```python
df2.columns.str.replace(" ", "_")
```

```python
df2.columns= df2.columns.str.replace("_", " ")
```

```python
#REPLACE() METHOD ....
df2.columns= df2.columns.str.replace(" ", "_")
df2.head(5)
```

```python
#CHANGING CASES FOR COLUMN LABELS....
df2.columns = [x.upper() for x in df2.columns]
df2.head(5) #
```

```python
#RENAME(mapper, axis, inplace) METHOD.....
df2.rename({"ROLL_NO":"roll no"}, axis=1, inplace=True)
df2.head(3)
```

```python
#ROW INDICES MODIFICATION....

# a) set_index(keys, drop, inplace)
df2.set_index(keys="roll no", drop=True, inplace=True)
df2
```

```python
# b) reset_index(inplace)
df2.reset_index(inplace=True)
df2
```

```python
#MODIFYING MULTIPLE ROWS/RECORDS IN DATA.....
# a) map(function, iterable(s))
pd.Series(map(len, df["name"])) #series obj containing length of names
```

```python
#Better Option for map()
df["name"].map(len) #also doing same task but without typecasting...
```

```python
#user defined function in map()...
def myFunc(x):
    if x >= 50:
        return "Old"
    else:
        return "Young"
df["age"].map(myFunc)
    
```

```python
#Adding new Column in data frame and result of the map function
df["Age Category"]=df["age"].map(myFunc)
```

```python
df
```

```python
#LAMBDA FUNCTION IN map()....
df["Age_Category"]=df["age"].map(lambda x: "Young" if x<50 else "Old")
df
```

```python
#dictionary in map....
df['session'].map({'MORNING':'M', 'AFTERNOON':'A'})
#limitation that where it doesnot fina a match it adds a NaN
```

```python
#  b) replace(to_replace, value, inplace=False)
df['session'].replace({'MORNING':'M', 'AFTERNOON':'A'}, inplace=True) #working properly no NAN Values..
```

```python
#replace() on entire dataframe.....
df.replace({"MOR":"M", "AFT":"A", "group A":"Group-A"})
```
