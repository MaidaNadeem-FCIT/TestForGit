```python
#RELPLOTS.....
# GENERIC SETTINGS...
sns.set_style(style='darkgrid') #STYLE ---> 'white' ,'dark', 'darkgrid', 'whitegrid'
sns.set_context(context='paper', font_scale=1.5) #CONTEXT --->'talk' ,'poster', 'paper'
```

```python
# 1. LINEPLOT....
sns.relplot(x='sepal_width', y='sepal_length', data=df_i, kind='line')
```

```python
# OTHER PARAMS OF RELPLOT...
sns.relplot(x='sepal_width', y='sepal_length', data=df_i, kind='line', hue='species')
```

```python
#SUBPLOTS..
sns.relplot(x='sepal_width', y='sepal_length', data=df_i, kind='line', col='species')
```

```python
sns.relplot(x='sepal_width', y='sepal_length', data=df_i, kind='line', hue='species', style='species')
```

```python
#   2. SCATTER PLOT....
sns.relplot(x='total_bill', y='tip', data=df_t ,kind='scatter')
```

```python
sns.relplot(x='total_bill', y='tip', data=df_t ,kind='scatter', hue='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=df_t ,kind='scatter', hue='day', style='day')
```

```python
sns.relplot(x='total_bill', y='tip', data=df_t ,kind='scatter', hue='sex', style='sex')
```

```python
#SUBPLOTS....
#as replot() is built on top of facetgrid() so, we can have subplots...
#TWO PLOTS IN SINGLE ROW...
sns.relplot(x='total_bill', y='tip', data=df_t, kind='scatter', col='smoker')
```

```python
#TWO PLOTS IN SINGLE COL...
sns.relplot(x='total_bill', y='tip', data=df_t, kind='scatter', row='smoker')
```

```python
#four PLOTS IN SINGLE ROW...
sns.relplot(x='total_bill', y='tip', data=df_t, kind='scatter', col='day')
```

```python
#four PLOTS IN 2 ROW 2 COLS...
sns.relplot(x='total_bill', y='tip', data=df_t, kind='scatter', col='day', col_wrap=2)
```
