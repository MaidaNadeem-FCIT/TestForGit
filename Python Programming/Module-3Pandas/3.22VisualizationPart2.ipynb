```python
#DATA VISUALIZATION PART 2.....
from matplotlib import pyplot as plt
import numpy as np
```

```python
# BAR-PLOT .....
students = ["Hadeed", "Maaz", "Mujahid", "Mohid", "Kamal",  "Jamal"]
english = [85, 45, 60, 70, 83, 35]

fig, ax= plt.subplots()
ax.bar(x=students, height=english, width=0.5)
ax.set_xlabel(" Student Names")
ax.set_ylabel("English Marks")

plt.title("Distribution of English Marks")
```

```python
# BAR-PLOT (More than One bar Plot).....
students = ["Hadeed", "Maaz", "Mujahid", "Mohid", "Kamal",  "Jamal"]
english = [85, 45, 60, 70, 83, 35]
maths=[69,78,48,85,65,90]

fig, ax= plt.subplots()
ax.bar(x=students, height=english, width=0.5)
ax.bar(x=students, height=maths, width=0.5)

ax.set_xlabel(" Student Names")
ax.set_ylabel("English Marks")

plt.title("Distribution of English Marks") #OVERLAPPED BARS CAUSING ISSUE
```

```python
# BAR-PLOT (More than One bar Plot without overlap).....
students = ["Hadeed", "Maaz", "Mujahid", "Mohid", "Kamal",  "Jamal"]
english = [85, 45, 60, 70, 83, 35]
maths=[69,78,48,85,65,80]

fig, ax= plt.subplots()
index=np.arange(len(students))
ax.bar(x=index, height=english, width=0.3, label="English")
ax.bar(x=index+0.3, height=maths, width=0.3, label="Maths")

ax.set_xlabel(" Student Names")
ax.set_ylabel("English Marks")

plt.xticks(ticks=index+0.15, labels=students, c='b')

plt.legend()
plt.title("Distribution of English Marks")
```

```python
#HORIZONTAL BAR GRAPH....
students = ["Hadeed", "Maaz", "Mujahid", "Mohid", "Kamal",  "Jamal"]
english = [85, 45, 60, 70, 83, 35]
maths=[69,78,48,85,65,80]

fig, ax= plt.subplots()
index=np.arange(len(students))
ax.barh(y=index, width=english, height=0.3, label="English")
ax.barh(y=index+0.3, width=maths, height=0.3, label="Maths")

ax.set_xlabel("Marks ")
ax.set_ylabel("Student Names")

plt.yticks(ticks=index+0.15, labels=students, c='b')

plt.legend()
plt.title("Distribution of English Marks")
```

```python
# REAL LIFE DATA SET....
import pandas as pd
df=pd.read_csv("datasets/so_survey_subset.csv")
df.shape
```

```python
df
```

```python
df.info() #we have alot of NaN values so need to handle them
```

```python
#  STEP--1 REMOVE NAN's
df1 = df.dropna() #easiest way is to drop NaN valu rows now we have a reduced data set
```

```python
df1.info() # no null values in dataframe any more...
```

```python
#  STEP--2 Languages Worked With Series to get
languages=df1['LanguageWorkedWith']
languages
```

```python
#  STEP-->3 Dictionary of Languages and their count.
from collections import Counter
l1=['C', 'Python', 'java', 'HTML']
myctr=Counter(l1)
myctr
```

```python
myctr.update(['Python', 'C'])
myctr
```

```python
ctr=Counter()
for lan in languages:
    ctr.update(lan.split(";"))
ctr #dictionary of languages nad their count in series as key : value pair
```

```python
mydict=dict(ctr)
languages=list(mydict.keys())
popularity=list(mydict.values())
```

```python
languages
```

```python
popularity
```

```python
# plot bar graph
fig=plt.figure()
ax=fig.add_subplot()

ax.barh(y=languages, width=popularity, height=0.6)
plt.tight_layout()
```
