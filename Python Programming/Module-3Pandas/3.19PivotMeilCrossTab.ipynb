```python
#2. ---> MELT(id_vars, value_vars, var_name, value_name)....
df=pd.read_csv('datasets/weather.csv')
df
```

```python
df1=df.melt(id_vars=['day'])
df1
```

```python
df.melt(id_vars=['day'], var_name='City', value_name='Temperature', value_vars=['karachi'])
```

```python
df.melt(id_vars=['day'], var_name='City', value_name='Temperature', value_vars=['lahore'])
```

```python
df.melt(id_vars=['day'], var_name='City', value_name='Temperature', value_vars=['murree'])
```

```python
df1[df1['variable']=='murree']
```

```python
df1[df1['variable']=='murree'].value.agg('mean')
```

```python
#3.---->CROSSTAB() for comparing categorical values...
df=pd.read_csv("datasets/sample1.csv")
df
```

```python
pd.crosstab(df.city, df.gender)
```

```python
pd.crosstab(index=df.city, columns=df.gender, margins=True)
```

```python
pd.crosstab(index=df.city, columns=df.gender, normalize=True) #% OF FREQUENCY
```

```python
import numpy as np
```

```python
pd.crosstab(index=df.city, columns=df.gender, values=df.age,aggfunc=np.mean)
```
