```python
#HANDLING REPEATING VALUES IN DATAFRAME.....
df=pd.read_csv("datasets/group-marks.csv")
df.head()
```

```python
df.session
```

```python
df.session.value_counts()
```

```python
#create a mapping dictionary to replace repeating values
dict1={
    'MORNING':'MOR',
    'MOR':'MOR',
    'AFTERNOON':'AFT',
    'AFT':'AFT',
    'EVENING':'EVE',
    'EVE':'EVE'
}
df.session.map(dict1)
```

```python
df['session']=df.session.map(dict1)
df
```

```python
df.session.value_counts()
```

```python
#CREATE NEW COLUMN USING AN EXISTING COLUMN.....
df['Scholarship_in_$']=df.scholarship.apply(lambda x: x/205)
df.head(10)
```

```python
#DELETING ROWS WITH NaN USING dropna(axis=0, how='any')
df.shape
```

```python
df1= df.dropna(axis=0, how='any') #drop a row if any value is NaN
df1.shape
```

```python
df2=df.dropna(how='all') #drop a row if all values are NaN
df2.shape #no row is NaN completely so no change....
```

```python
#subset arguement...
df3=df.dropna(axis=0, how='any', subset=['math'])
df3.shape
```

```python
df4=df.dropna(axis=0, how='all', subset=['math', 'session'])
df4.shape
```

```python
df5=df.dropna(axis=0, how='any', subset=['math', 'session']) #Drop a row if any value in math or session is NaN
df5.shape
```

```python
df6=df.dropna(axis=1, how='all')  #drop a column if all values in that column is NaN
df6.head()#no column in dropped 
```

```python
df7=df.dropna(axis=1, how='any')  #drop a column if all values in that column is NaN
df7.head()#Three columns are dropped  here
```
