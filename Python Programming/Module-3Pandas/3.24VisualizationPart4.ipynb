```python
#BOX PLOT....
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("datasets/boxplot.csv")
df.head()
```

```python
# CALCULATE REQUIRED ARGUMMENTS...
df.data_science.quantile([0.25, 0.50, 0.75])
```

```python
#PLOT OF boxplot()....
fig,ax = plt.subplots()
ml=list(df['machine_learning'])
ds=list(df['data_science'])
bd=list(df['big_data'])
values=[ml, ds, bd]

ax.boxplot(x=values, patch_artist=True, labels=["Machine Learning", "Data Science", "Big Data"])
ax.set_xlabel("Subjects")
ax.set_ylabel("Marks")
ax.set_title("BoxPlot for Subject Marks")
ax.yaxis.grid(True)
```

```python
#VIOLIN PLOT....
fig,ax = plt.subplots()
ml=list(df['machine_learning'])
ds=list(df['data_science'])
bd=list(df['big_data'])
values=[ml, ds, bd]
labels=["Machine Learning", "Data Science", "Big Data"]


ax.violinplot(dataset=values, showmedians=True, quantiles=[[0.25,0.75],[0.25,0.75] ,[0.25,0.75]])
ax.set_xticks([1,2,3])
ax.set_xticklabels(labels)

ax.set_xlabel("Subjects")
ax.set_ylabel("Marks")
ax.set_title("BoxPlot for Subject Marks")
ax.yaxis.grid(True)
```

```python
#HEAT MAP....
# used to show relationship between two variables
# visualize 2D data using colors...

data=np.random.randint(0,255,100).reshape(10,10)
data
```

```python
plt.imshow(X=data)
plt.colorbar()
plt.title("2D Heat Map")
```

```python
plt.imshow(X=data, cmap='gray')
plt.colorbar()
plt.title("2D Heat Map")
```

```python
plt.imshow(X=data, cmap='coolwarm')
plt.colorbar()
plt.title("2D Heat Map")
```
