```python
#MERGING JOINING APPENDING CONCATENATE....
import pandas as pd
```

```python
# PART 1 ---> Merging and Joining.....
df_temp=pd.DataFrame({
    "city":["Lahore", "Murree", "Peshawar","Sialkot"],
    "Temperature":[39,14,29,32]
})
df_temp
```

```python
df_humid=pd.DataFrame({
    "city":["Karachi","Lahore","Peshawar","Lahore","Murree"],
    "humidity":[76,95,72,70,75]
})
df_humid
```

```python
# 1->INNER JOIN (Intersection)
d1=pd.merge(df_temp,df_humid,on='city', how='inner', indicator=True)
d1
```

```python
d2=pd.merge(df_humid,df_temp,on='city', how='inner', indicator=True)
d2
```

```python
# 2--->OUTER JOIN (Union/Full Join)..
d3=pd.merge(df_temp, df_humid, on='city', how='outer', indicator=True)
d3
```

```python
d4=pd.merge(df_humid, df_temp, on='city', how='outer', indicator=True)
d4
```

```python
# 3.--->LEFT JOIN...
d5=pd.merge(df_temp, df_humid, on='city', how='left', indicator=True)
d5
```

```python
d6=pd.merge(df_humid, df_temp, on='city', how='left', indicator=True)
d6
```

```python
# 4.---->RIGHT JOIN(Right outer join)...
d7=pd.merge(df_temp, df_humid, on='city', how='right', indicator=True)
d7
```

```python
d8=pd.merge(df_humid, df_temp, on='city', how='right', indicator=True)
d8
```

```python
# SUFFIXES PARAMETER....
df1=pd.DataFrame({
    'city':['Lahore','Karachi','Murree'],
    'Temperature':[29,31,15],
    'humidity':[69,89,56]
})
df2=pd.DataFrame({
    'city':['Karachi','Lahore','Islamabad'],
    'Temperature':[32,39,16],
    'humidity':[91,92,67]
})
df1
```

```python
df2
```

```python
d0 = pd.merge(df1,df2, on='city', how='inner', suffixes=('_left', '_right')) # will add suffixes to the columns of approprite df
d0
```

```python
#PART II ----> Concatenate and Append...
# 1. ROW WISE WITH SAME COULUMNS..
pw = pd.DataFrame({
    'city': [ 'Lahore', 'Karachi', 'Peshawer', 'Islamabad', 'Muree'],
    'temperature' : [35, 39, 33, 29, 15],
    'humidity' : [76, 95, 72, 81, 70],
})
pw
```

```python
uw = pd.DataFrame({
    'city': [ 'Dubai', 'Sharja', 'Ajman', 'Abu Dhabi'],
    'temperature' : [41, 44, 47, 45],
    'humidity' : [88, 99, 79, 86],
})
uw
```

```python
dd1=pd.concat([pw,uw],axis=0) #concatenate to dataframes vertically but indices are not good do 
dd1
```

```python
dd1=pd.concat([pw,uw],axis=0, ignore_index=True) #Now Indices are in order and unique 
dd1
```

```python
# creating sub groups in concatenated dataframe 
dd2=pd.concat([pw,uw],axis=0, keys=['Pak','UAE']) #Now DataFrame is divided in two groups of Respective cities of countries 
dd2
```

```python
dd2.loc['Pak']
```

```python
dd2.loc['UAE']
```
