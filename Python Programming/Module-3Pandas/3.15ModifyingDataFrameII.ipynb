```python
# 9. SORTING 
df = pd.DataFrame({
    'roll_no': [ 102, 101, 104, 103, 105],
    'name' : ['Kamal', 'Saima', 'Jamal', 'Shaikh', 'Farzana'],
    'gender' : ['M', 'F', 'M', 'M', 'F'],
    'grade'  : ['A', 'A', 'B', 'B', 'A'],
    'marks'  : [ 21,  23,  12,  14,  20],
    'city' : ['Lahore', 'Peshawer', 'Lahore', 'Karachi', 'Peshawer']
})
df
```

```python
# A) sort_values(by,axis=0, ascending=False, inplace=False, kind='quicksort', na_posiion='last',ignore_index=False)
df.sort_values(by=['grade']) #sort by grade but marks with 23 is at 2nd position 
```

```python
df2=df.sort_values(by=['grade','marks'], ascending=[True, False]) #multile columns based sorting and each with diff order
df2 #indeices stay with record and randomized
```

```python
df2.reset_index(drop=True, inplace=True) #as we want index to stay in particular order we reset them 
df2 #and drop old indices and save permanently
```
