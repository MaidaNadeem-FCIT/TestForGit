```python
#SELECTION/SUBSETTING....
df=pd.read_csv("datasets/groupdata.csv")
df.set_index(keys='roll no', drop=True, inplace=True)
df
```

```python
dfs=df.sort_values("age") #now indices are out of order
dfs
```

```python
#SUBCRIPT......
dfs[['name']].head(10) #return series object with Name Column having top 10 values
```

```python
dfs[['name', 'age']].head(10)
```

```python
dfs['MS14':'MS10':] #upper limit included as string indices in slice object of subscript
#but for integer actual indices upper limit is exclusive
```

```python

dfs[1:2:1]# only 1 row at positiona index 1 will be selected
```

```python
dfs[::3] #from 0 to end with step 3
```

```python
dfs["MS09":"MS11":-2]
```

```python
#ILOC[rowstoselect, colstoselect]
#rows/cols to select can be --> scalar, list, slice object...
#stop value in slice object will be excluded here as iloc[] works on positional indices and execute naturally with indices..
dfs
```

```python
dfs.iloc[2,:] #row at position 2 with all cols
```

```python
dfs.iloc[[2,4,1],:]
```

```python
dfs.iloc[2:6,:]
```

```python
dfs.columns
```

```python
#selecting columns....
dfs.iloc[:, 4] #all rows of column at 4th position -->group
```

```python
dfs.iloc[:, [1,3,5]] #columns at 1st, 3rd anf 5th position ---> (age, session , gender)
```

```python
dfs.iloc[:, 3:6] #all cols from 3 to 5 as upper limit is exclusive
```

```python
#select rows+cols...
dfs.iloc[[2,4,0], 2:6]#--->2nd, 4th and 0th row  + 2nd to 5th cols
```

```python
#LOC[rows to select, cols to select]  ----> works on actual indices and stop value in inclusive in slice object
dfs.loc['MS03':'MS16', ['age','group']] #UPPER LIMIT IS INCLUDED
```

```python
dfs.loc[['MS07','MS11','MS03'],:]
```

```python
dfs.loc['MS02':'MS05':-1,:] #reverse slicing
```

```python
dfs.loc['MS02':'MS05':-1,'group':'age':-1] #reverse slicing
```

```python
#CONDITIONAL SELECTION....
#OPTION 1
l1=[]
for leng in df.age:
    if leng>40:
        l1.append(True)
    else:
        l1.append(False)
l1
```

```python
#OPTION 2
df[df.age>40]
```
